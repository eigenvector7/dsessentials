```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def neg(x):return -x
def pwr(x,y,c):return c*(x**y)/math.factorial(y)
L=range(1,8)
k=range(0,20)
for l in L:
    const=np.exp(-l)
    pr=list(map(pwr,[l]*len(k),k,[const]*len(k)))
    plt.plot(k,pr,'--*',label="L={}".format(l))
    plt.legend()
plt.show()
```
